## 1.Problem Statement:

"In the rapidly evolving telecommunications industry, customer churn and dissatisfaction have become pressing concerns, adversely affecting telecom revenue. This problem statement seeks to investigate the root causes and repercussions of customer churn and dissatisfaction in the telecommunications sector, with the goal of identifying strategies and solutions to mitigate their impact on telecom revenue."

**Effect on Telecom Revenue:**

* The high rate of customer churn in the telecom industry is leading to a significant loss in subscription revenue, as customers switch to competing service providers.
* Increasing customer dissatisfaction with service quality, pricing, and customer support is resulting in lower customer lifetime value, reduced upsell opportunities, and diminished revenue potential.
* The proliferation of negative online reviews, social media complaints, and word-of-mouth disapproval is tarnishing the brand image of telecom companies, discouraging potential customers, and causing a decline in new customer acquisition and retention efforts.
* Inefficient billing practices, hidden fees, and inconsistent service quality are contributing to an elevated number of billing disputes, leading to revenue leakage and costly dispute resolution processes.
* The inability to effectively anticipate and address customer needs and preferences is causing missed opportunities for targeted marketing campaigns, personalized upselling, and cross-selling, which could otherwise boost average revenue per user (ARPU).
* Escalating customer service costs, including call center operations and handling of service inquiries and complaints, are eroding the profit margins of telecom companies, diminishing their overall revenue and profitability.

To address these challenges, telecom companies must conduct a comprehensive analysis of the factors contributing to customer churn and dissatisfaction, implement customer-centric strategies, and leverage advanced technologies and data analytics to enhance service quality, reduce churn, and optimize revenue-generation opportunities.

## 6. Data Transformation and Preprocessing

```python
cleaned_df = df.drop(['UpdatedAt','customerID'],axis =1)# dropping UpdatedAt and Customer ID columns
```

```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
for col in cleaned_df:
    if cleaned_df[col].dtype == 'object':
        # If 2 or fewer unique categories
        if len(list(cleaned_df[col].unique())) == 2:
            # Train on the training data
            lb.fit(cleaned_df[col])
            # Transform both training and testing data
            cleaned_df[col] = lb.transform(cleaned_df[col])
```

```python
cleaned_df.shape
```

```python
from sklearn.preprocessing import MinMaxScaler# using MinMax Scaller as low number or outliers and rest of the data encoded using lable encoder
```

```python
scaler = MinMaxScaler()
X_Scaledata = scaler.fit_transform(cleaned_df)
```

```python
Scaled_Data = pd.DataFrame(data=X_Scaledata, columns=["gender", "SeniorCitizen","Partner","tenure","PhoneService","StreamingTV","InternetService","PaperlessBilling","MonthlyCharges","TotalCharges","Churn"])
Scaled_Data
```

 ### 1.Create Logistic regression model

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score
def Models_Evaluation_Metrics (model,x_rtrain,y_rtrain,x_test,y_test,y_test_pred): 
    acc_train=round(model.score(x_rtrain,y_rtrain)*100,2)
    acc_test=round(model.score(x_test,y_test)*100,2)
    roc_auc= round(roc_auc_score(y_test, y_test_pred),4)

    return acc_train, acc_test, roc_auc
    
    print("Training Accuracy: {}%".format(acc_train))
    print("Test Accuracy: {}%".format(acc_test))
    print('ROC AUC:', roc_auc)
```

```python
from sklearn.linear_model import LogisticRegression
LOR_model = LogisticRegression()
```

```python
LOR_model.fit(X_rtrain,y_rtrain)
```

```python
LOR_ypred = LOR_model.predict(X_test)
```

```python
LOR_accuracy_score = accuracy_score(y_test,LOR_ypred)*100
LOR_confusion_matrix = confusion_matrix(y_test,LOR_ypred)
LOR_classification_report = classification_report(y_test,LOR_ypred)
print(f'Logistic model accuracy is {LOR_accuracy_score}')
print(LOR_classification_report)
print(f'Logistic model confusion matrix is\n {LOR_confusion_matrix}')
      
```

```python
sns.heatmap(LOR_confusion_matrix,annot =True,fmt = 'd')
```

```python
ndf = [Models_Evaluation_Metrics(LOR_model,X_rtrain,y_rtrain,X_test,y_test,LOR_ypred)]

logmodel_score = pd.DataFrame(data = ndf, columns=['Training Accuracy','Test Accuracy','ROC AUC'])
logmodel_score.insert(0, 'Model', 'Linear Regression')
```
